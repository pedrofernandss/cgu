```python
import pandas as pd
```

```python
# Conversor de arquivos de excel (.xlsx) para csv

excel_url = ""
df = pd.read_excel(excel_url)

csv_url = ""
df.to_csv("", index=False)

print("Convertido!")
```

```python
# Conversor de arquivos de csv para parquet

csv_url = ""
df = pd.read_csv(csv_url)

parquet_url = ""
df.to_parquet(parquet_url)

print("Convertido!")
```

```python
# Conversor de arquivo de parquet para excel (.xlsx)

parquet_url = ""
df = pd.read_parquet(parquet_url)

excel_url = ""
df.to_excel(excel_url, index=False)

print("Convertido!")
```
