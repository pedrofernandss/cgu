```python
import pandas as pd
```

```python
tcesDatabaseURL = '../database/raw/tces_raw.xlsx'
tcesDataframe = pd.read_excel(tcesDatabaseURL)
tcesDataframe.info()
```

```python
colunasDeInteresse = ['Ministério', 'Ano da instauração do processo pela CGU', 'ano_provavel_contrato',
                       'gestao', 'governo_tce', 'alinhamento_ano_contrato', 'alinhamento_tce', 'Motivo da Instauração da TCE', 'name_region']

tcesDataframe = tcesDataframe[colunasDeInteresse]
```

```python
novosNomesDasColunas = {
    'Ano da instauração do processo pela CGU': 'ano_instauracao_processo_cgu',
    'Motivo da Instauração da TCE': 'motivo_instauracao_tce',
    'name_region': 'regiao'
}

tcesDataframe.rename(columns=novosNomesDasColunas, inplace=True)
```

```python
# Remoção dos acrónimos e o traço/barra
tcesDataframe['Ministério'] = tcesDataframe['Ministério'].str.replace(r'^.*[-/]\s*', '', regex=True)

# Remoção de espaços extras no início/fim e normalizar
tcesDataframe['Ministério'] = tcesDataframe['Ministério'].str.strip()

# Correção de erros de formatação (ex: 'Ministérioda Justiça' -> 'Ministério da Justiça')
tcesDataframe['Ministério'] = tcesDataframe['Ministério'].str.replace(r'Ministérioda', 'Ministério da', regex=True)

# Converção de tudo para a primeira letra maiúscula e o resto minúsculo
tcesDataframe['Ministério'] = tcesDataframe['Ministério'].str.title()
```

```python
localSalvamento = '../database/preprocessed/tces_processado.xlsx'
tcesDataframe.to_excel(localSalvamento, index=False)
```
