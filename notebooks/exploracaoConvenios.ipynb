```python
conveniosDataframe["DATA FINAL VIGÊNCIA"] = pd.to_datetime(conveniosDataframe["DATA FINAL VIGÊNCIA"], dayfirst=True)
conveniosDataframe["ano_final_vigencia"] = conveniosDataframe["DATA FINAL VIGÊNCIA"].dt.year
```

```python
colunasDeInteresse = conveniosDataframe[['NÚMERO CONVÊNIO', 'UF', 'NOME ÓRGÃO SUPERIOR', 'valor_convenio_defla', 'ano_final_vigencia','alinhamento.gov', 'alinhamento.final']]
colunasDeInteresse
```
